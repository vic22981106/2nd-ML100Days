```python
import os
import numpy as np
import pandas as pd
```

```python
# setting the data_path
dir_data = './Day2data/'
path = os.path.join(dir_data, 'application_train.csv')
print('Path of my data: %s' % (path))
df = pd.read_csv(path)
```

```python
# print the first five rows of data in the file("application_train.csv")
df.head()
```

```python
# using the df.shape method
# the output will be list containing number of rows and number of columns
# for example (5,3) 5 represents the number of rows in a specific csv file and 3 indicates that there are 3 columns in this file
df.shape
# in this case, the output is (30751,122), which means that there are 307511 rows and 122 columns in this csv file !!!
```

```python
# Q: How can i get the whole name of a specific column ?
# Ans: using the "columns" method, the output will be a list containing the names of entire columns in this file
print(df.columns)
print("\n")

# you can also print the names of the specific columns in this file by using index 
# for example i print the names of the first 10 columns
print(df.columns[0:10])
```

```python
# How do i get the data from specfic columns?
# as we can get the name of the specific columns
# we can just access the data from a specific columns by its name
# the following is an example, i want to know the data from the "AMT_INCOME_TOTAL" column
df.AMT_INCOME_TOTAL
```

```python
# here is another example, when i want to know the data from the "NAME_CONTRACT_TYPE" column
df.NAME_CONTRACT_TYPE
```

```python
# there is an important thing we have to take into consideration when we analyse the data
# that is if there is any empty value in the specific row and column?
# so in order to find the empty value, we can use the method, isnull() and notnull()
path_1 = os.path.join(dir_data, 'GoldenStatesWarriors.csv')
df_1 = pd.read_csv(path_1)
print(df_1)
print("\n")
print(df_1.loc[:, "Health"].isnull()) #using loc method we can access the data from specific row and column
# False indicates that it is not empty in that location
print("\n")
print(df_1.loc[:, "Points"].notnull())
# True indicates that it is not empty in that location 
```
