```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate x, y data
a = np.random.randint(0,1000)
np.random.seed(a)
x = np.random.randint(0,50,1000)
y = -2*x + np.random.normal(0,20,1000)

# calculate the correlation coefficient
np.corrcoef(x, y)
```

```python
# plot the scattering plot
plt.scatter(x, y)
```
