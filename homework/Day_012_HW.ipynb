```python
# import those package i need
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

利用pandas.cut()這一個函式可以將數值進行分割
比如說可以自訂一個分類boundaries，如下面這一個cell所示：
bins_cut是一個我自訂的年齡間距
所以如果利用pd.cut(ages['age'], bins=bins_cut,labels=False)這一個函式，就可以將['age']這一個column依據我設計的bins_cut來進行分類
所以執行之後，一共會分成5組 ：(0, 10], (10, 20], (20, 30], (30, 50], (50, 100]。

right 如果是True，分出來的分類boundary就會包含上界，但不包含下界，如果是False，就會是相反的。
labels這一個功能如果設定成False，就會將(0, 10], (10, 20], (20, 30], (30, 50], (50, 100]這五組，分成0,1,2,3,4這5組
retbins 如果是True，就會把bins return回來，如果bins是scalar type是比較useful的 ( default = False)
precision (default = 3) 會控制最後呈現的boundaries的精準度，也就是小數點下第幾位的準確度
include_lowest : Whether the first interval should be left-inclusive or not.

```python
# generate a data frame  'age'
ages = pd.DataFrame({"age": [18,22,25,27,7,21,23,37,30,61,45,41,9,18,80,100]})

bins_cut = (0,10,20,30,50,100)
ages['customized_age_grp'] = pd.cut(ages['age'],bins = bins_cut,right = True)
ages['customized_age_grp_type'] = pd.cut(ages['age'],bins = bins_cut,right = True,labels=False)
ages.head()
```

也可以針對數值的分佈進行分類
比如說: pd.cut(ages['ages'], bins = 4)會將ages這一個dataFrame裡面其中'ages'這一個column裡面的數值分成四類，分類的依據是將年分成齡四等份，所以每一組的差距間隔是固定的，可以看下面這一個cell的output，每一組的年齡兼具是固定的，所以每一組的成員數量都不一樣。

```python
ages["equal_width_age"] = pd.cut(ages["age"], bins=4,retbins=False)
ages['equal_width_type'] = pd.cut(ages["age"], bins=4, labels=False)
ages['equal_width_age'].value_counts()
```

另外也可以根據數量進行分類，比如說pd.qcut(ages['age'],bins = 4)這一個函式就是將age這一個column的數值依照大小排序分成四類，每一組的數量都是相同的，所以可以看到下面的output，每一組的間距都不同，但是數量都一樣。

```python
ages["equal_freq_age"] = pd.qcut(ages["age"],4)
ages['equal_freq_type']= pd.qcut(ages["age"],4,labels=False)
ages['equal_freq_age'].value_counts()
```
